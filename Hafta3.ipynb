```python
#  kullanılacak sayıyı iste
değişken = int(input('3 haneli bir doğal sayı giriniz.'))

#  sayı kendisinden küçük rakamlara bölünebilir mi?
gruplayıcı = list( range(2,min(9,değişken)+1) )

print( '0\'a bölüm tanımsızdır.' )
print( 'Her sayı 1\'e tam bölünür.' )
for üye in gruplayıcı:
  if değişken % üye == 0:
    print( değişken, üye, '\'ye tam bölünür.')
```

```python
#  İçlem ile liste yaratmak
#  250'yi tam bölen rakamlar
[
    üye for üye in gruplayıcı
    if değişken % üye == 0
]
```

```python
len( [i for i in gruplayıcı if değişken%i==0] )
```
