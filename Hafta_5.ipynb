#  Boolean Masking: Veri filtreleme

```python
auto
```

```python
#   2000 kilodan hafif araçları listele:
#   weight < 2000

auto[ auto.weight < 2000 ]
```

```python
auto[ auto.weight < 2000 ].describe().T
```

```python
#  1800 < weight < 2000

auto[ (auto.weight > 1800) & (auto.weight < 2000) ]
```

```python
auto.origin.unique()
```

```python
#  Meneşi 2 olan araçlar
auto[ auto.origin == 2 ]
```

```python
#  Markası "vw" veya "volkswagen" olanlar

#    markasında vw geçenler
[ 'vw' in i for i in auto.name ]
```

```python
auto[
    [ ('vw' in i) or ('volkswagen ' in i) for i in auto.name ]
]
```
