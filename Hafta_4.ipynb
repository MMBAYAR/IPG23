```python
kat1 = []
kat2 = []
kat3 = []
kat4 = []

for i in range(10**7):
  d = numpy.random.choice(
      ['4000\'den az','4001-9000','9001-14000','14ooo\'den çok'],
      p = [20/1000,210/1000,325/1000,445/1000]
  )

  if d == ['4000\'den az','4001-9000','9001-14000','14ooo\'den çok'][0]:
    kat1.append(1)
  elif d == ['4000\'den az','4001-9000','9001-14000','14ooo\'den çok'][1]:
    kat2.append(1)
  elif d == ['4000\'den az','4001-9000','9001-14000','14ooo\'den çok'][2]:
    kat3.append(1)
  else:
    kat4.append(1)
```

```python
#  Sayaçlar listesini oluştur
sayaçlar = []
for i in range(10**5):
  #  Sayaç ile koleksiyon kaçıncı paket alımında tamamlandı izle
  sayaç = 0
  #  Önce 7 tane faklı sonuç birikine kadar deney yap
  biriken = []
  while len( set(biriken) ) < 7:
    sayaç += 1
    #    rassal olay
    d = numpy.random.choice(
        [1,2,3,4,5,6,7],
        p = [4/27,6/27,7/27,1/27,3/27,2/27,4/27]
        )
    biriken.append(d)

  sayaçlar.append(sayaç)
```
