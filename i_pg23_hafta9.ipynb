##  **Bir Veri Tablosu Oluşturulması**<br><br>
Veri tabloları, önceden satır ve sütun sayısı belirlenerek ve isteğe bağlı olarak içine yazılacak veri tanımlanarak oluşturulur.<br><br>
4 satırı ve 1 sütunu olan bir veri tablosu oluşturup bu veri tablosunu df adındaki bir değişkene atayalım:
<br><br>
Bunun için pandas'ın ".DataFrame()" fonksiyonunu kullanacağız. "D" ve "F"nin büyük harfle yazıldığına lütfen dikkat ediniz.<br><br>
Bu fonksiyonun iki tane olmazsa olmaz argümanı vardır:
- index,
- columns

**index**, satırları adreslemekte kullanacağınız indislerin bir gruplayıcı biçiminde sunulmasını gerektirir. Gruplayıcının içerdiği üye sayısı ile veri tablonuzun satır sayısını da belirlemiş olursunuz.<br><br>
**columns**, sütunları adreslemekte kullanacağınız sütun başlıklarının bir gruplayıcı biçiminde sunulmasını gerektirir. Gruplayıcının içerdiği üye sayısı ile veri tablonuzun sütun sayısını da belirlemiş olursunuz.
<br><br>

```python
#  Bir veri tablosu oluşturmak
#    - 4 satırlı,
#    - 1 sütunlu bir veri tablosu olacak

#  adı "df" olan bir değişken tanımlıyorum
df = None
#  boş bir veri tablosu atıyorum
df = pd.DataFrame(
    index = ['İlk','İkinci','Üçüncü','Dördüncü'],
    columns = ['Ad']
)
```

```python
#  Oluşturduğum Veri Tablosunu inceliyorum
df
```

<br><br>
"**NaN**": "Not a number" demektir, "None" değerinin pandastaki karşılığıdır. Boş hücrelerde görülür.<br><br>
##  **Veri Tablosuna Veri Girmek**<br><br>
Bir veri tablosu için "**.iloc[...]**" metodu, veri tablosunun istenilen satırını satır sırasına göre getirir.
<br><br>

```python
#  .iloc[] yöntemi ile değer girmek: ilk satırda verimiz 'Mariama' olsun.

df.iloc[ 0 ] = ['Mariama']
df
```

```python
#  Veri tablosunun ilk satırını çağırmak

df.iloc[0]
```

<br><br>
Bir veri tablosu için "**.loc[...]**" metodu, veri tablosunun istenilen satırını indis adına göre getirir.
<br><br>

```python
#  .loc[] yöntemi ile değer girmek: "İkinci" satırının verisi 'Burak' olsun.

df.loc[ 'İkinci' ] = ['Burak']
df
```

```python
#  Veri tablosunun istenilen satırlarını çağırmak:


df.loc[ ['İlk','İkinci'] ]
```

<br><br>
"**.loc[ indis_adı ]**" metodu ile veri girişi, tabloda **henüz var olmayan** bir indis için uygulandığında tabloya o indis adıyla yeni bir satır ekler.
<br><br>

```python
df.loc['Önceden_Olmayan_Yeni_İndis'] = [None for i in df.columns]
df
```

<br><br>
Neden bu yeni satıra veri girmek için "df.columns gruplayıcısının tüm üyeleri için..." yapısının kullanıldığını düşününüz.
<br><br>

<br><br>
##  **Veri Tablosuna Yeni Bir Sütun Eklemek**
<br><br>
Bir veri tablosuna yeni bir sütun eklemek için önce bir veri tablosunun istenilen bir sütun verisini çağırmayı görelim.<br><br>
"**df[ sütun_adı ]**"<br><br>ya da<br><br>
"**df.sütun_adı**"<br><br>
kullanımı, ilgili sütun verisini çağıracaktır. Ancak; sütun adları boşuk karakteri içeriyorsa ikinci yöntem çalışmayacaktır. Alışkanlık olarak ilk yöntemi kullanmanız önerilir.<br><br>

```python
#  Ad sütunundaki veriyi çağırıyorum

df['Ad']
```

```python
#  Ad sütununa veri giriyorum:

df['Ad'] = ['Mariama', 'Burak', 'Gül', 'Ayşe Nur', 'Bir diğer öğrenci']
df
```

<br><br>
Sütun adı anarak veri girişi yapmak, veri tablosunda **henüz var olmayan** bir sütun adı ile uygulandığında o ada sahip yeni bir sütun ekleyecektir.
<br><br>

```python
df['Yeni sütun'] = [None for i in df.index]
df
```

<br><br>
Neden veri girişinde "df.index gruplayıcısının tüm üyeleri için..." veri tanımlaması yapıldığını düşününüz.
<br><br>

```python
#  df.sütun_adı metodu ile sütun verisi çağırmanın kırılganlığı:

df.Yeni sütun
```

```python
#  df.sütun_adı metodu, yeni sütun eklemek için de kullanılamaz!
df.yeni_sütun = [None for i in df.index]
df
```

##  **Bir Veri Tablosundan Bir Sütunu ya da Satırı Silmek**<br><br>

İlgilendiğimiz veri tablosunun "*değişken*" adındaki bir değişkene atandığını varsayarak başlayalım.<br><br>
***değişken.drop( satır/sütun_adı, axis = 0/1 )***<br><br>
taslak kodunda olduğu gibi **.drop()** metodu ile veri tablolarından satırlar ya da sütunlar silinebilir.
- ***değişken.drop( satır_adı, axis = 0 )*** örneğinde görüldüğü gibi ilgili veri tablosundan bir satır silinmek istenildiğinde **.drop()** metodu içine satırın adı (indisi) ve ***axis = 0*** parametresi girilmelidir.
- ***değişken.drop( sütun_adı, axis = 1 )*** örneğinde görüldüğü gibi ilgili veri tablosundan bir sütun silinmek istenildiğinde **.drop()** metodu içine sütunun adı (başlığı) ve ***axis = 1*** parametresi girilmelidir.

<br>Silinecek satırın adını (indisini) ya da sütunun adını (başlığını) hatırlayamıyor ancak; bu satırın ya da sütunun sırasını hatırlıyorsanız son eklenen satır adı için;<br><br>
***değişken.drop( df.index[-1], axis = 0 )***<br><br>
ve son eklenen sütun adı için de;<br><br>
***değişken.drop( df.columns[-1], axis = 1 )***<br><br>
kod taslaklarında sunulan gruplayıcı içinden son üyeyi getirme mantığı örnek alınabilir.<br><br>


```python
#  Veri tablomuzun silme işlemlerinden önceki halini hatırlayalım
df
```

```python
#  Veri tablosuna eklenen son sütunu silelim

df.drop(
    ['Yeni sütun',],
    axis = 1
)
```

```python
#  Yeniden veri tablomuzu inceleyelim

df
```

<br><br>
**DİKKAT!: Silme işlemleri sonunda elde edilen yeni veri tablosu eski veri tablosunu içeren değişkenin yeni değeri olarak güncellenmediği sürece orijinal veri tablosu korunacaktır.**
<br><br>

```python
#  Veri tablomuzu son sütunu silinmiş hali olarak güncelleyelim

df = df.drop( df.columns[-1], axis = 1)

df
```

<br><br>
**Şimdi bir de son satırı da silelim...**
<br><br>

```python
#  Veri tablomuzu son satırı silinmiş hali olarak güncelleyelim

df = df.drop( df.index[-1], axis = 0 )

df
```

<br><br>
Arzu edilen işlem; yalnızca yanlış girilmiş veya değeri değişmiş bir verinin değiştirilmesi ise bütün bir satırın ya da sütunun silinmesine gerek olmayabilir.<br><br>
Bu tür durumlarda yalnızca ilgili veri tablosu hücresinin değerini yeniden tanımlamak yeterli olacaktır.
<br><br>

```python
#  Öğrencilerin Harf Notunu İçeren Bir Sütunu Veri Tablomuza Ekleyelim

from numpy import random

df['Harf_Notu'] = random.choice( ['AA','BB','CC','DD','FF'], size = 4, p = [1/5,1/5,1/5,1/5,1/5,] )

df
```

```python
#  Şimdi veri tablosunda ilk öğrencinin notunun "AA" yerine başka bir not girildiğini düşünelim

#    Yanlış girilmiş olan notu görmenin bazı farklı yolları
df.iloc[0,1], df['Harf_Notu'][0], df.Harf_Notu[0], df.loc['İlk', 'Harf_Notu']
```

```python
#    Biz "Mariama"nın notunu yanlış girdiğimizi biliyoruz ancak; "Mariama"nın veri tablomuzda kaçıncı öğrenci olduğunu hatırlayamıyorsak;

#      Önce liste/sözlük içlemlerinde öğrendiğimiz filtreleme yaklaşımını veri tablolarında deneyelim:
#      Veri tablosunun Ad'ı 'Mariama' olan alt kümesini çağıralım

df[ df.Ad == 'Mariama' ]
```

```python
list(df.index).index('İlk')
```

```python
#      Şimdi de bu alt kümede 'Harf_Notu'nu düzeltelim

#        .loc[] metodu ile veri güncelleyelim
df.loc[ df[ df.Ad == 'Mariama' ].index, 'Harf_Notu' ] = 'AA'

#        .iloc[] metodu ile veri güncelleyelim (Ad değeri 'Mariama' olan satır indisi, indisler listesinin kaçıncı elemanıdır?)
df.iloc[ list(df.index).index(df[ df.Ad == 'Mariama' ].index), 1 ] = 'AA'

df
```

##  **İki Veri Tablosunun Birleştirilmesinde .append() Metodu Kullanımı**
